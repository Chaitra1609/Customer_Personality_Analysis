```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python

# File path of the Excel file
file_path = "E:/Important project/marketing_campaign.xlsx"

# Load the cleaned sheet into a DataFrame
df_cleaned = pd.read_excel(file_path, sheet_name="Cleaned sheet", engine="openpyxl")

# Display the DataFrame
print(df_cleaned.head())

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15,5))

# Income Boxplot
plt.subplot(1,3,1)
sns.boxplot(x=df['Income'])
plt.title("Income Outliers")

# Recency Boxplot
plt.subplot(1,3,2)
sns.boxplot(x=df['Recency'])
plt.title("Recency Outliers")

# Total Spending Boxplot
plt.subplot(1,3,3)
sns.boxplot(x=df['Total Spending'])
plt.title("Total Spending Outliers")

plt.show()
```

```python
Q1 = df['Recency'].quantile(0.25)
Q3 = df['Recency'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Recency Outlier Range: Less than {lower_bound} or more than {upper_bound}")

# Check if there are any outliers
outliers_recency = df[(df['Recency'] < lower_bound) | (df['Recency'] > upper_bound)]
print(outliers_recency)
```

```python
df[df['Recency'] > 149]
```

```python
Q1_income = df['Income'].quantile(0.25)
Q3_income = df['Income'].quantile(0.75)
IQR_income = Q3_income - Q1_income

lower_bound_income = Q1_income - 1.5 * IQR_income
upper_bound_income = Q3_income + 1.5 * IQR_income

print(f"Income Outlier Range: Less than {lower_bound_income} or more than {upper_bound_income}")
```

```python
df[(df['Income'] < lower_bound_income) | (df['Income'] > upper_bound_income)]
```

```python
Q1_spending = df['Total Spending'].quantile(0.25)
Q3_spending = df['Total Spending'].quantile(0.75)
IQR_spending = Q3_spending - Q1_spending

lower_bound_spending = Q1_spending - 1.5 * IQR_spending
upper_bound_spending = Q3_spending + 1.5 * IQR_spending

print(f"Total Spending Outlier Range: Less than {lower_bound_spending} or more than {upper_bound_spending}")
```

```python
df[(df['Total Spending'] < lower_bound_spending) | (df['Total Spending'] > upper_bound_spending)]
```

```python
# Remove outliers for Income
df = df[~((df['Income'] < lower_bound_income) | (df['Income'] > upper_bound_income))]

# Remove outliers for Total Spending
df = df[~((df['Total Spending'] < lower_bound_spending) | (df['Total Spending'] > upper_bound_spending))]

# Reset the index after removal
df = df.reset_index(drop=True)

print(f"Dataset after removing outliers: {df.shape}")
```

```python
# Set the figure size
plt.figure(figsize=(12, 5))

# Boxplot for Income
plt.subplot(1, 2, 1)
sns.boxplot(y=df['Income'])
plt.title("Boxplot for Income")

# Boxplot for Total Spending
plt.subplot(1, 2, 2)
sns.boxplot(y=df['Total Spending'])
plt.title("Boxplot for Total Spending")

plt.show()
```

# Correlation Analysis

```python
#Correlation matrix to identify strong relationships:
numerical_df = df.select_dtypes(include=['int64', 'float64'])
```

```python
corr_matrix = numerical_df.corr()
strong_corr = corr_matrix[(corr_matrix > 0.5) | (corr_matrix < -0.5)]
print(strong_corr)
```

```python
import numpy as np

# Calculate full correlation matrix
corr_matrix = numerical_df.corr()

# Get upper triangle of correlation matrix (to avoid duplicate pairs)
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Find features with correlation > 0.85
high_corr_pairs = [
    (col, row, upper_tri.loc[row, col])
    for col in upper_tri.columns
    for row in upper_tri.index
    if abs(upper_tri.loc[row, col]) > 0.85
]

# Display them
for col1, col2, corr_val in high_corr_pairs:
    print(f"{col1} <---> {col2} | Correlation: {corr_val:.2f}")
```

```python
# Correlation with 'Total Spending'
spending_corr = numerical_df.corr()['Total Spending'].sort_values(ascending=False)

# Display top correlations
print(spending_corr)
```
